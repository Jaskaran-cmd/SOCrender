```python
discharge_data=discharge_data.drop(['SerialNum','TempNum','MaxV_No','MinV_No','MaxT_No','MinT_No','CellV_1',
       'CellV_2', 'CellV_3', 'CellV_4', 'CellV_5', 'CellV_6', 'CellV_7',
       'CellV_8', 'CellV_9', 'CellV_10', 'CellV_11', 'CellV_12', 'CellV_13',
       'CellV_14', 'CellV_15', 'Temp_1', 'ErrCode1', 'ErrCode2', 'ErrCode3',
       'ErrCode4', 'ErrCode5', 'ErrCode6', 'ErrCode7', 'ErrCode8',
       'DI1', 'DI2', 'DI3', 'DI4', 'DO1', 'DO2', 'DO3', 'DO4', 'BoardNum'], axis=1)
discharge_data.shape
```

```python
discharge_data=discharge_data.drop(['CycleTime'],axis=1)
```

```python
fig = px.line(discharge_data, x="Time", y="MinV")
fig.show()
```

```python
discharge_data.Time
```

```python
discharge_data['Hours']=discharge_data['Time'].dt.hour
discharge_data['Minute']=discharge_data['Time'].dt.minute
```

```python
discharge_data['TimeM']=(discharge_data.Hours-15)*60+discharge_data.Minute-39
discharge_data=discharge_data.drop(['Hours','Minute','Time'],axis=1)
discharge_data.TimeM.unique()
```

```python
discharge_data.info()
```

```python
discharge_data=discharge_data.drop(['ChgMOS','DisChgMOS'],axis=1)
```

```python
X=discharge_data[['Bms_Life','SumVoltage','Current','RemainCap','MaxV','MinV','MaxT','MinT','TimeM']].values
y=discharge_data['SOC'].values
```

```python
discharge_data.info()
```

```python
discharge_data=discharge_data.drop(['Bms_Life'],axis=1)
```

```python
train_size=int(0.8*len(discharge_data))
train_X,train_y=X[:train_size],y[:train_size]
test_X,test_y=X[train_size:],y[train_size:]

# train_X = torch.from_numpy(train_X).type(torch.Tensor)
# train_y = torch.from_numpy(train_y).type(torch.Tensor)
# test_X = torch.from_numpy(test_X).type(torch.Tensor)
# test_y = torch.from_numpy(test_y).type(torch.Tensor)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

Model2=LinearRegression()
Model2.fit(train_X,train_y)
res=Model2.predict(test_X)
r2=r2_score(test_y,res)
print( r2*100)
```

```python
import pickle
trained_model="trained_model.sav"
pickle.dump(Model2,open(trained_model,'wb'))

```
